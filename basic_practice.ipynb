```python
import sklearn
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer(as_frame = True)
cancer_df = cancer.frame
cancer_target = cancer_df['target']
cancer_features = cancer_df.drop(columns ='target')
cancer_features.boxplot(vert = False)
```

```python
import matplotlib.pyplot as plt
ax = cancer_df.plot.scatter(x = 'worst area', y = 'worst concave points', c = 'target', colormap = 'bwr', s = 2)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cancer_features, cancer_target, random_state = 11)
```

```python
# need to scale data for algo w. distance based measure
from sklearn.preprocessing import StandardScaler
import pandas as pd
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns = cancer_df.columns[:-1])
X_train_scaled_df.head(5)
```
